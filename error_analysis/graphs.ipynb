```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

file = Path().absolute()
df = pd.read_csv(file / "csvs" / "variation_in_K.csv")
```

```python
plt.figure(figsize=(8, 5))

for matrix, group in df.groupby("matrix"):
    # Average across replicates if there are multiple rows per K
    avg = group.groupby("K", as_index=False)["frob_norm"].mean()

    plt.plot(
        avg["K"],
        avg["frob_norm"],
        marker="o",
        label=matrix
    )

plt.xlabel("K")
plt.ylabel("Average Frobenius Norm")
plt.title("Subspace Estimation Error vs K")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
